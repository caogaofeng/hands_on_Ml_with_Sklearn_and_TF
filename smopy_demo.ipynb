```python

hz = smopy.Map((32.073190, -113.397964, 42.208454, -125.388431))
#hz = smopy.Map((30.0, 119.8, 30.55, 120.5),z=10) z指定缩放程度
#hz.show_mpl()

print(housing["longitude"].mean())
housing["longitude"].fillna(housing["longitude"].mean(),inplace=True)
housing["latitude"].fillna(housing["latitude"].mean(),inplace=True)


print('longitude:', housing["longitude"][0], "latitude", housing["latitude"][0])
#第一个点40.063883, -121.872615
#第二个个点37.924938, -120.395925
x1 = [40.063883, 37.924938]
y1 = [-121.872615, -120.395925]
#x, y = hz.to_pixels(pd.Series(x1), pd.Series(y1))
#x, y = hz.to_pixels(data["latitude"][0], data["longitude"][0])
#x, y = hz.to_pixels(30.25, 37.88)
#ax.scatter(x, y)
#print("x=", x, "y=", y)
ax = hz.show_ipython(figsize=(12, 15))


for i in range(len(housing["latitude"])):
    x, y = hz.to_pixels(housing["latitude"][i], housing["longitude"][i])
    print(i)
    print("x=", x, "y=", y)
    ax.scatter(x, y, alpha=0.4, s=housing['population'][i]/100, label='population',
            c=housing['median_house_value'][i], cmap=plt.get_cmap('jet'),
            edgecolors='black')

plt.show()
```
